## Small World Graphs

A small-world network refers to an ensemble of networks in which the mean geodesic (i.e., shortest-path) distance between nodes increases sufficiently slowly as a function of the number of nodes in the network (http://www.scholarpedia.org/article/Small-world_network). <br>
The Watts‚ÄìStrogatz model is a random graph generation model that produces graphs with small-world properties, including short average path lengths and high clustering. We use Watts-Strogatz generation for the following experiments. <br>
NetworkX (https://networkx.org/documentation/networkx-1.9/reference/generated/networkx.generators.random_graphs.watts_strogatz_graph.html) has a function 'watts_strogatz_graph' that returns a Watts-Strogatz small-world graph. The input parameters are n (the number of nodes), k (each node is connected to k nearest neighbors in ring topology), and p (probability of rewiring each edge). <br>
Here is a brief description of the Watts-Strogatz algorithm: First create a ring over $n$ nodes. Then each node in the ring is joined to its ùëò nearest neighbors (or $k$‚àí1 neighbors if $k$ is odd). Then shortcuts are created by replacing some edges as follows: for each edge ($u$,$v$) in the underlying ‚Äú$n$-ring with $k$ nearest neighbors‚Äù with probability $p$ replace it with a new edge ($u$,$w$) with uniformly random choice of existing node $w$.

```python
sw = nx.generators.random_graphs.watts_strogatz_graph(20, 20, 0.1)
nx.draw(sw)
```

### Experiment 1

```python
g1, = plt.plot(x1,y1)
g2, = plt.plot(x2,y2)
g3, = plt.plot(x3,y3)
g4, = plt.plot(x4,y4)
g5, = plt.plot(x5,y5)
g6, = plt.plot(x6,y6)
g7, = plt.plot(x7,y7)
g8, = plt.plot(x8,y8)
g9, = plt.plot(x9,y9)
plt.legend([g1, g2, g3, g4, g5, g6, g7, g8, g9], ('p=0.1','p=0.2','p=0.3','p=0.4','p=0.5','p=0.6','p=0.7','p=0.8','p=0.9'))
```

From the graph above, it seems that the actual probability of rewiring does not affect the efficiency of the minimum spanning tree of the small-world graph. However, we see that increasing clustering increases the efficiency of the small-world graph minimum spanning tree. This seems intuitive enough. However, I imagine that when robustness is measured, robustness will decrease as clustering increases (https://www.nature.com/articles/srep37317).

## Scale-Free Graphs

Generally a network is deemed scale free if the fraction of nodes with degree $k$ follows a power-law distribution $k^{-\alpha}$, where $\alpha < 1$. 

### Barab√°si-Albert Graph
A graph of $n$ nodes is grown by attaching new nodes each with $m$ edges that are preferentially attached to existing nodes with high degree.

```python
sf = nx.generators.random_graphs.barabasi_albert_graph(20,1)
nx.draw(sf)
```

#### Experiment 1
How does the efficiency of the minimum spanning tree of a Barab√°si-Albert graph change with an increase in $m$ as the number of nodes $n$ stays constant?

```python
x1 = []
y1 = []
for i in range(1, 50):
    x1.append(i)
    sw = nx.generators.random_graphs.barabasi_albert_graph(50, i)
    mst = nx.minimum_spanning_tree(sw)
    y1.append(nx.global_efficiency(mst))
plt.plot(x1, y1)
```

The efficiency of the minimum spanning trees of the graphs increases as the value of $m$ increases.
